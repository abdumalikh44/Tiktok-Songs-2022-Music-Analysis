#2. Data Cleaning

```python
#ganti nama kolom
df.columns = ['trc_nm', 'art_nm', 'art_pp', 'alb', 'trc_pp', 'dncblt','enrg', 'ldns', 'mode','key', 'spchs', 'acstc', 'instr', 'lvns', 'vlnc', 'tmp', 'tm_sg', 'drt_ms']
```

```python
df.columns
```

```python
#Cek value apakah tetap konstan dari 0.0 ke 1.0 (danceability)
df = df[(df["dncblt"] >= 0.0) & (df["dncblt"] <= 1.0)]
print(df.to_string())
```

```python
#Cek value apakah tetap konstan dari 0.0 ke 1.0 (energy)
df = df[(df["enrg"] >= 0.0) & (df["enrg"] <= 1.0)]
print(df.to_string())
```

```python
#Cek value apakah tetap konstan dari 0.0 ke 1.0 (acousticness)
df = df[(df["acstc"] >= 0.0) & (df["acstc"] <= 1.0)]
print(df.to_string())
```

```python
#Cek value apakah tetap konstan dari 0.0 ke 1.0 (valence)
df = df[(df["vlnc"] >= 0.0) & (df["vlnc"] <= 1.0)]
print(df.to_string())
```

```python
#Cek value apakah tetap konstan dari -60 ke 0 (loudness)
df = df[(df["ldns"] >= -60) & (df["ldns"] <= 0)]
print(df.to_string())
```

```python
#Cek value apakah tetap konstan dari 3 ke 7 (time_signature )
df = df[(df["tm_sg"] >= 3) & (df["tm_sg"] <= 7)]
print(df.to_string())
```

```python
#Cek value apakah hanya 1 atau 0 (mode)
df = df[(df["mode"] == 1) | (df["mode"] == 0)]
print(df.to_string())
```

```python
#Cek value apakah tetap konstan dari 0 ke 100 (artist_pop)
df = df[(df["art_pp"] >= 0) & (df["art_pp"] <= 100)]
print(df.to_string())
```

```python
#Cek value apakah tetap konstan dari 0 ke 100 (track_pop)
df = df[(df["trc_pp"] >= 0) & (df["trc_pp"] <= 100)]
print(df.to_string())
```

```python
#Remove all duplicates
df.drop_duplicates(inplace = True)
print(df.to_string())
```

##Membuat Heatmap
##Melihat besaran koefisien korelasi antara tiap atribut predictor dengan atribut target

```python
# Compute the correlation matrix
correlation_matrix = df.corr()
#Graph I.
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title("Correlation Matrix Heatmap")
plt.show()


#Graph II
# Create a heatmap of the correlations with the target column
corr = df.corr()
target_corr = corr['trc_pp'].drop('trc_pp')

# Sort correlation values in descending order
target_corr_sorted = target_corr.sort_values(ascending=False)

sns.set(font_scale=0.8)
sns.set_style("white")
sns.set_palette("PuBuGn_d")
sns.heatmap(target_corr_sorted.to_frame(), cmap="coolwarm", annot=True, fmt='.2f')
plt.title('Correlation with trcak_pop')
plt.show()
```
